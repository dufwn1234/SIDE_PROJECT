```python
import pandas as pd
import requests
from datetime import datetime
import time
```

```python
def beauty(key,api_call_limit,page_now,indx,api_call, character_beauty=None):
    headers = {
    "x-nxopen-api-key": key
    }
    
    now = datetime.now()
    today = str(now)[:8] + str(int(str(now)[8:10])-1)
    
    if character_beauty is None:
        character_beauty = pd.DataFrame()
    
    for page in range(page_now,100):
        now_page = page
        urlString = f"https://open.api.nexon.com/maplestory/v1/ranking/overall?date={today}&world_type=0&page={page}"
        response = requests.get(urlString, headers = headers)
        api_call += 1
        if api_call == api_call_limit:
            return character_beauty.reset_index(drop=True), api_call, now_page,0
        
        a =  response.json()
        ranking = pd.json_normalize(a['ranking'])
        
        character_name = list(ranking['character_name'])
        
        for i,j in enumerate(character_name[indx:]):
            urlString_2 = "https://open.api.nexon.com/maplestory/v1/id?character_name=" + j
            response_ocid = requests.get(urlString_2, headers = headers)
            ocid = response_ocid.json()
            api_call += 1
            if api_call == api_call_limit:
                return character_beauty.reset_index(drop=True), api_call, now_page, i+indx
            
            time.sleep(1)
            urlString_3 = "https://open.api.nexon.com/maplestory/v1/character/basic?ocid=" + ocid['ocid']
            response_basic = requests.get(urlString_3, headers = headers)
            api_call += 1
            if api_call == api_call_limit:
                return character_beauty.reset_index(drop=True), api_call, now_page, i+indx

            character_basic = response_basic.json()
            character_basic = pd.json_normalize(character_basic)[['character_name', 'world_name', 'character_gender', 'character_class','character_level','character_guild_name']]
            
            time.sleep(1)
            urlString_4 = "https://open.api.nexon.com/maplestory/v1/character/beauty-equipment?ocid=" + ocid['ocid']
            response_beauty = requests.get(urlString_4, headers = headers)
            api_call += 1
            if api_call == api_call_limit:
                return character_beauty.reset_index(drop=True), api_call, now_page, i+indx

            character = response_beauty.json()
            if str(character['character_class']) == '제로' or str(character['character_class']) == '엔젤릭버스터':
                hair = pd.json_normalize(character['character_hair'])
                hair.rename(columns={'base_color':'hair_base_color','mix_color':'hair_mix_color','mix_rate':'hair_mix_rate'},inplace=True)
                
                face = pd.json_normalize(character['character_face'])
                face.rename(columns={'base_color':'face_base_color','mix_color':'face_mix_color','mix_rate':'face_mix_rate'},inplace=True)
                
                skin = pd.DataFrame({'skin_name':  [character['character_skin_name']]})
                if i == 0 and page == page_now:
                    character_beauty = pd.concat([character_basic, hair,face,skin], axis = 1)
                else:
                    b = pd.concat([character_basic, hair,face,skin], axis = 1)
                    character_beauty = pd.concat([character_beauty, b], axis = 0)
            
                additional_hair = pd.json_normalize(character['additional_character_hair'])
                additional_hair.rename(columns={'base_color':'hair_base_color','mix_color':'hair_mix_color','mix_rate':'hair_mix_rate'},inplace=True)
                
                additional_face = pd.json_normalize(character['additional_character_face'])
                additional_face.rename(columns={'base_color':'face_base_color','mix_color':'face_mix_color','mix_rate':'face_mix_rate'},inplace=True)

                additional_skin = pd.DataFrame({'skin_name':  [character['additional_character_skin_name']]})

                b = pd.concat([character_basic, additional_hair, additional_face, additional_skin], axis = 1)
                character_beauty = pd.concat([character_beauty, b], axis = 0)

            else:
                hair = pd.json_normalize(character['character_hair'])
                hair.rename(columns={'base_color':'hair_base_color','mix_color':'hair_mix_color','mix_rate':'hair_mix_rate'},inplace=True)
                
                face = pd.json_normalize(character['character_face'])
                face.rename(columns={'base_color':'face_base_color','mix_color':'face_mix_color','mix_rate':'face_mix_rate'},inplace=True)
                
                skin = pd.DataFrame({'skin_name':  [character['character_skin_name']]})
                if i == 0 and page == page_now:
                    character_beauty = pd.concat([character_basic, hair,face,skin], axis = 1)
                else:
                    b = pd.concat([character_basic, hair,face,skin], axis = 1)
                    character_beauty = pd.concat([character_beauty, b], axis = 0)
        indx = 0
```

```python
key = ""
character_beauty_1,api_call,page_now,idx = beauty(key,1000,1,0,0)
```

```python
character_beauty_1.to_csv('character_beauty_1.csv',index=False)
```

```python
def cashitem(key,api_call_limit,page_now,indx,api_call, character_cashitem_equipment=None):
    headers = {
    "x-nxopen-api-key": key
    }
    
    now = datetime.now()
    today = str(now)[:8] + str(int(str(now)[8:10])-1)
    
    if character_cashitem_equipment is None:
        character_cashitem_equipment = pd.DataFrame()
    
    for page in range(page_now,100):
        now_page = page
        urlString = f"https://open.api.nexon.com/maplestory/v1/ranking/overall?date={today}&world_type=0&page={page}"
        response = requests.get(urlString, headers = headers)
        api_call += 1
        if api_call == api_call_limit:
            return character_cashitem_equipment.reset_index(drop=True), api_call, now_page, 0
        
        a =  response.json()
        ranking = pd.json_normalize(a['ranking'])
        
        character_name = list(ranking['character_name'])
        
        for i,j in enumerate(character_name[indx:]):
            urlString_2 = "https://open.api.nexon.com/maplestory/v1/id?character_name=" + j
            response_ocid = requests.get(urlString_2, headers = headers)
            ocid = response_ocid.json()
            api_call += 1
            if api_call == api_call_limit:
                return character_cashitem_equipment.reset_index(drop=True), api_call, now_page, i+indx
            time.sleep(1)
            
            urlString_3 = "https://open.api.nexon.com/maplestory/v1/character/basic?ocid=" + ocid['ocid']
            response_basic = requests.get(urlString_3, headers = headers)
            api_call += 1
            if api_call == api_call_limit:
                return character_cashitem_equipment.reset_index(drop=True), api_call, now_page, i+indx

            character_basic = response_basic.json()
            character_basic = pd.json_normalize(character_basic)[['character_name', 'world_name', 'character_gender', 'character_class','character_level','character_guild_name']]

            time.sleep(1)
            urlString_4 = "https://open.api.nexon.com/maplestory/v1/character/cashitem-equipment?ocid=" + ocid['ocid']
            response_cashitem = requests.get(urlString_4, headers = headers)
            api_call += 1
            if api_call == api_call_limit:
                return character_cashitem_equipment.reset_index(drop=True), api_call, now_page, i+indx

            cashitem_equipment = response_cashitem.json()
            preset_no = cashitem_equipment['preset_no']
            
            if preset_no is None or cashitem_equipment[f'cash_item_equipment_preset_{preset_no}'] == []:
                if str(character_basic['character_class']) == '제로' or str(character_basic['character_class']) == '엔젤릭버스터':
                    if i == 0 and page == page_now:
                        cashitem_equipment_preset = pd.json_normalize(cashitem_equipment['cash_item_equipment_base'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                        character_cashitem_equipment = pd.merge(character_basic, cashitem_equipment_preset, how='cross')
                    else:
                        cashitem_equipment_preset = pd.json_normalize(cashitem_equipment['cash_item_equipment_base'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                        character_cashitem_equipment_preset = pd.merge(character_basic, cashitem_equipment_preset, how='cross')
                        character_cashitem_equipment = pd.concat([character_cashitem_equipment, character_cashitem_equipment_preset], axis = 0)
                
                    additional_cashitem_equipment_preset = pd.json_normalize(cashitem_equipment['additional_cash_item_equipment_base'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                    character_additional_cashitem_equipment_preset = pd.merge(character_basic, additional_cashitem_equipment_preset, how='cross')
                    character_cashitem_equipment = pd.concat([character_cashitem_equipment, character_additional_cashitem_equipment_preset], axis = 0)
                    
                else:
                    if i == 0 and page == page_now:
                        cashitem_equipment_preset = pd.json_normalize(cashitem_equipment['cash_item_equipment_base'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                        character_cashitem_equipment = pd.merge(character_basic, cashitem_equipment_preset, how='cross')
                    else:
                        cashitem_equipment_preset = pd.json_normalize(cashitem_equipment['cash_item_equipment_base'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                        character_cashitem_equipment_preset = pd.merge(character_basic, cashitem_equipment_preset, how='cross')
                        character_cashitem_equipment = pd.concat([character_cashitem_equipment, character_cashitem_equipment_preset], axis = 0)
            else:
                if str(character_basic['character_class']) == '제로' or str(character_basic['character_class']) == '엔젤릭버스터':
                    if i == 0 and page == page_now:
                        cashitem_equipment_preset = pd.json_normalize(cashitem_equipment[f'cash_item_equipment_preset_{preset_no}'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                        character_cashitem_equipment = pd.merge(character_basic, cashitem_equipment_preset, how='cross')
                    else:
                        cashitem_equipment_preset = pd.json_normalize(cashitem_equipment[f'cash_item_equipment_preset_{preset_no}'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                        character_cashitem_equipment_preset = pd.merge(character_basic, cashitem_equipment_preset, how='cross')
                        character_cashitem_equipment = pd.concat([character_cashitem_equipment, character_cashitem_equipment_preset], axis = 0)
                
                    additional_cashitem_equipment_preset = pd.json_normalize(cashitem_equipment[f'additional_cash_item_equipment_preset_{preset_no}'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                    character_additional_cashitem_equipment_preset = pd.merge(character_basic, additional_cashitem_equipment_preset, how='cross')
                    character_cashitem_equipment = pd.concat([character_cashitem_equipment, character_additional_cashitem_equipment_preset], axis = 0)
                    
                else:
                    if i == 0 and page == page_now:
                        cashitem_equipment_preset = pd.json_normalize(cashitem_equipment[f'cash_item_equipment_preset_{preset_no}'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                        character_cashitem_equipment = pd.merge(character_basic, cashitem_equipment_preset, how='cross')
                    else:
                        cashitem_equipment_preset = pd.json_normalize(cashitem_equipment[f'cash_item_equipment_preset_{preset_no}'])[['cash_item_equipment_part','cash_item_name','cash_item_label']]
                        character_cashitem_equipment_preset = pd.merge(character_basic, cashitem_equipment_preset, how='cross')
                        character_cashitem_equipment = pd.concat([character_cashitem_equipment, character_cashitem_equipment_preset], axis = 0)
        indx = 0
```
